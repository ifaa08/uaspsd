*   train_test_split dari sklearn.model_selection: Digunakan untuk membagi dataset menjadi dua bagian, yaitu data latih (training data) dan data uji (test data).

```python
# Import library
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
import pickle
```

**membaca data file csv**

*   pd.read_csv('data.csv'): Ini adalah panggilan fungsi dari pandas untuk membaca data dari file CSV. Fungsi read_csv ini membaca data dari file CSV dan mengonversinya menjadi suatu struktur data yang disebut DataFrame.

'data.csv': Parameter pertama berisi nama file CSV yang akan dibaca.
*   data = merupakan variabel yang menyimpan Hasil dari fungsi pd.read_csv disimpan dalam variabel yang dinamakan data. Variabel ini akan berisi DataFrame yang memuat seluruh data dari file CSV.



```python
# Membaca data dari file CSV
data = pd.read_csv('data.csv')
```

**memisahkan fitur dan target**

*   X =Hasil dari operasi pemilihan kolom tersebut disimpan dalam variabel X. Variabel X akan berisi DataFrame yang hanya terdiri dari fitur-fitur yang telah dipilih.
*   data[['process.b1.capacity', 'process.b2.capacity', 'process.b3.capacity', 'process.b4.capacity', 'property.price', 'property.product', 'property.winner']]:
Ini adalah operasi pemilihan kolom pada DataFrame data. Dalam konteks ini, kita memilih beberapa kolom sebagai fitur untuk model kita.
Fitur-fitur yang dipilih diambil dari kolom-kolom dengan nama 'process.b1.capacity', 'process.b2.capacity', 'process.b3.capacity', 'process.b4.capacity', 'property.price', 'property.product', dan 'property.winner'.
*   y =Hasil dari operasi pemilihan kolom target disimpan dalam variabel y. Variabel y akan berisi Series yang berisi nilai dari kolom 'verification.result'.
*   data['verification.result']: Ini adalah operasi pemilihan kolom untuk memilih kolom 'verification.result', yang merupakan target atau label dari model.





```python
# Memisahkan fitur-fitur dan target
X = data[['process.b1.capacity', 'process.b2.capacity', 'process.b3.capacity', 'process.b4.capacity', 'property.price', 'property.product', 'property.winner']]
y = data['verification.result']
```

**membagi data menjadi training dan testing set**

*   train_test_split(X, y, test_size=0.1, random_state=42):

Fungsi ini mengambil empat argumen:
1. X: Fitur-fitur yang ingin dibagi (dalam hal ini, DataFrame X yang berisi fitur-fitur).
2. y: Target atau label yang ingin dibagi (dalam hal ini, Series y yang berisi label).
3. test_size: Proporsi data yang akan dialokasikan sebagai set pengujian. Dalam contoh ini, test_size=0.1 berarti 10% dari data akan menjadi set pengujian, sedangkan 90% akan menjadi set pelatihan.
4. random_state: Digunakan untuk memastikan hasil yang reprodusibel. Angka apa pun dapat digunakan sebagai nilai ini.
*   X_train, X_test, y_train, y_test =

Variabel X_train dan y_train akan berisi set pelatihan untuk fitur dan target.
Variabel X_test dan y_test akan berisi set pengujian untuk fitur dan target.
Setelah menjalankan baris kode ini, kita memiliki empat variabel yang memuat data terbagi, yaitu:

X_train: Fitur dari set pelatihan.
X_test: Fitur dari set pengujian.
y_train: Target dari set pelatihan.
y_test: Target dari set pengujian.



```python
# Membagi data menjadi training dan testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

**standarisasi fitur**
*   scaler = StandardScaler(): Ini adalah pembuatan objek StandardScaler. StandardScaler digunakan untuk melakukan penskalaan fitur sehingga setiap fitur memiliki rata-rata nol dan deviasi standar satu. Proses ini berguna untuk mengatasi perbedaan skala antar fitur, yang dapat memengaruhi performa beberapa model pembelajaran mesin.
*   nnX_test = scaler.fit(X_test): Metode fit dari StandardScaler digunakan untuk menghitung rata-rata dan deviasi standar dari set pengujian (X_test). Hasilnya disimpan dalam variabel nnX_test.



```python
# Standarisasi fitur
scaler = StandardScaler()
nnX_test = scaler.fit(X_test)
```

**Membangun dan melatih model MLPClassifier**

MLPClassifier adalah model yang dapat digunakan untuk tugas klasifikasi dengan memanfaatkan kekuatan dari jaringan saraf tiruan.MLPClassifier (Multi-Layer Perceptron Classifier) adalah model klasifikasi yang termasuk dalam kategori jaringan saraf tiruan (neural networks) dan disediakan oleh pustaka scikit-learn di Python.

-(Hidden Layers): MLPClassifier dapat memiliki satu atau lebih lapisan tersembunyi. Jumlah dan ukuran neuron dalam setiap lapisan dapat dikontrol oleh parameter hidden_layer_sizes. hidden_layer_sizes=(100, 50, 25)
Ada tiga lapisan tersembunyi di jaringan saraf:
- Lapisan pertama memiliki 100 unit atau neuron.

- Lapisan kedua memiliki 50 unit atau neuron.

- Lapisan ketiga memiliki 25 unit atau neuron.

Setiap unit dalam lapisan tersembunyi berperan sebagai pembangun blok untuk pembelajaran pola yang lebih kompleks dari data pelatihan. Jumlah unit dalam lapisan tersembunyi adalah salah satu faktor yang dapat mempengaruhi kapasitas model, di mana model dengan lebih banyak unit mungkin dapat mempelajari pola yang lebih rumit, tetapi juga memerlukan lebih banyak data dan waktu pelatihan.

- max_iter=1000: Ini menentukan jumlah iterasi maksimum atau epoch yang akan dilakukan selama pelatihan model. Dalam hal ini, pelatihan akan berhenti setelah 1000 iterasi jika konvergensi belum tercapai.

- activation='relu': Ini menentukan fungsi aktivasi yang digunakan di setiap unit atau neuron. 'relu' mengacu pada fungsi aktivasi ReLU (Rectified Linear Unit), yang umum digunakan dalam lapisan tersembunyi.

- random_state=42: Parameter ini memberikan nilai seed untuk menghasilkan hasil yang dapat direproduksi, seperti yang telah dijelaskan sebelumnya.

mlp.fit(X_train, y_train): Ini adalah langkah pelatihan model. Metode fit digunakan untuk melatih model dengan menggunakan data pelatihan.

- X_train: Subset fitur untuk pelatihan model.

- y_train: Subset target untuk pelatihan model.

Setelah pemanggilan fit, model akan belajar dari data pelatihan dan dapat digunakan untuk membuat prediksi pada data baru.

```python
# Membangun dan melatih model MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(100, 50, 25), max_iter=1000, activation='relu', random_state=42)
mlp.fit(X_train, y_train)
```

**Evaluasi model pada data pengujian**

1. accuracy = mlp.score(X_test, y_test):

- mlp: Merujuk ke objek model MLPClassifier yang telah dilatih sebelumnya.

- score(...): Metode score digunakan untuk mengukur akurasi model pada dataset yang diberikan. Akurasi diukur sebagai rasio jumlah prediksi yang benar terhadap jumlah total data.

- X_test: Subset fitur dari data pengujian.

- y_test: Subset target dari data pengujian.

Hasil dari pemanggilan score disimpan dalam variabel accuracy.

2. print(f'Accuracy: {accuracy * 100:.2f}%'):

- f'Accuracy: {accuracy * 100:.2f}%': Ini adalah f-string, yang digunakan untuk memasukkan nilai variabel (accuracy) ke dalam string. Angka setelah titik dua (:.2f) mengontrol jumlah digit desimal yang ditampilkan.

- accuracy * 100: Mengubah nilai akurasi menjadi persentase.

- :.2f: Menunjukkan dua digit desimal.

- %: Menambahkan simbol persentase.

Hasil akhirnya adalah mencetak akurasi model pada dataset pengujian ke konsol dengan format "Accuracy: xx.xx%".

```python
# Evaluasi model pada data pengujian
accuracy = mlp.score(X_test, y_test)
print(f'Accuracy: {accuracy * 100:.2f}%')
```
