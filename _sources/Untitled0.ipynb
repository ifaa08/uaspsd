```python
import streamlit as st
import pickle
import pandas as pd
import numpy as np

# Memuat objek StandardScaler dan model MLPClassifier dari file pickle
with open('scaler.pkl', 'rb') as scaler_file:
    scaler = pickle.load(scaler_file)

with open('mlp_model.pkl', 'rb') as model_file:
    mlp = pickle.load(model_file)

# Fungsi untuk melakukan prediksi
def predict_verification_result(data):
    # Mengubah data masukan pengguna menjadi DataFrame
    input_data = pd.DataFrame([data], columns=['process.b1.capacity', 'process.b2.capacity', 'process.b3.capacity', 'process.b4.capacity', 'property.price', 'property.product', 'property.winner'])

    # Normalisasi data menggunakan objek StandardScaler yang telah di-load
    input_data = scaler.transform(input_data)

    # Melakukan prediksi dengan model MLPClassifier
    prediction = mlp.predict(input_data)

    return prediction[0]

# Judul aplikasi Streamlit
st.title('Klasifikasi Hasil Verifikasi hasil lelang')

# Input form for new data
st.write('Data ini berisi tentang mencapai target pelelangan, apakah penawar memenangkan product dengan harga tersebut atau tidak (di tempat lelang).')
st.write('Form pengisian dibawah ini memiliki 7 kriteria yang harus diisi yaitu : 3 kriteria umum (property.price, property.product, dan property.winner) dan 4 kriteria lainnya (process.b1.capacity, process.b2.capacity, process.b3.capacity, process.b4.capacity).')
st.write('Sebelum mengisi form dibawah, diharapkan anda telah membaca deskripsi data yang harus diisikan pada setiap form, agar aplikasi dapat melakukan prediksi dengan optimal.')
# Create a button to trigger the modal
button_clicked = st.button('Baca Deskripsi Data Disini')

# Check if the button is clicked
if button_clicked:
    # Display modal content using markdown
    st.write('Kenapa data ini bisa disebut penting untuk para penawar? karna dari hasil klasifikasi ini para penawar bisa memperkirakan hal-hal yang perlu disiapkan untuk mencapai target pelelangan.')
    st.write('fitur ini berisi tentang penawaran sebuah product dari penawar 1- penawar 4,yang berisi 2043 data. dengan parameter kemenangan : harga,produk dan berapa kali penawar memenangkan pelelangan.')
    st.write('tujuan dari data ini adalah para penawar bisa memperkirakaan kemenangan pelelangan dengan parameter kemenangan dilihat dari harga,produk dan berapa kali penawar memenangkan pelelangan.')
    st.write('Data ini memverifikasi apakah harga tertentu mungkin untuk produk tertentu, dan apakah pemenang lelang tertentu kemungkinan besar memenangkan produk pada harga tersebut.Terdapat 7 fitur yaitu :')
    st.write('1. process.b1.capacity (penawar 1)')
    st.write('2. process.b2.capacity (penawar 2)')
    st.write('3. process.b3.capacity (penawar 3)')
    st.write('4. process.b1.capacity (penawar 4)')
    st.write('5. property.price (harga)')
    st.write('6. property.product (produk)')
    st.write('7. property.winner (pemenang)')
    st.button('Tutup Deskripsi')  # Optionally, add a button to close the modal

# Input data dari pengguna
st.write('Masukkan data berikut untuk mengklasifikasikan hasil verifikasi hasil lelang:')
process_b1_capacity = st.number_input('process.b1.capacity', value=0.0)
process_b2_capacity = st.number_input('process.b2.capacity', value=0.0)
process_b3_capacity = st.number_input('process.b3.capacity', value=0.0)
process_b4_capacity = st.number_input('process.b4.capacity', value=0.0)
property_price = st.number_input('property.price', value=0.0)
property_product = st.number_input('property.product', value=0.0)
property_winner = st.number_input('property.winner', value=0.0)

# Tombol untuk melakukan prediksi
if st.button('Prediksi Hasil Verifikasi'):
    # Membuat data masukan dari input pengguna
    user_input = {
        'process.b1.capacity': process_b1_capacity,
        'process.b2.capacity': process_b2_capacity,
        'process.b3.capacity': process_b3_capacity,
        'process.b4.capacity': process_b4_capacity,
        'property.price': property_price,
        'property.product': property_product,
        'property.winner': property_winner
    }

    # Melakukan prediksi
    prediction = predict_verification_result(user_input)

    # Menampilkan hasil prediksi
    st.write(f'Hasil Verifikasi: {prediction}')
```
